```python
#Câu 4:
mydict = {"a": 1, "b": 2, "c": 3, "d": 4}
```

```python
#Câu 4.A:
for key in mydict:
    print(key)
```

```python
#Câu 6a:
text = "jabbawocky"
vowels = {'a', 'e', 'i', 'o', 'u'}

sum = 0

for i in text:
    if i not in vowels:
        sum += 1
        
print("Số phụ âm nguyên âm:", sum)
```

```python
#Câu 6b:
text = "jabbawocky"
vowels = {'a', 'e', 'i', 'o', 'u'}

sum = 0

for i in text:
    if i in vowels:
        continue
    sum += 1
    
print("Số phụ âm nguyên âm:", sum)
```

```python
#Câu 7:
for i in range(-2,3):
    try:
        result = 10/i
        print("10/",i,"=",result)
    except ZeroDivisionError:
        print("Lỗi không chia hết cho 0!!!")
```
