```python
#Câu 2:
for i in range(11):
    if i%2 != 0:
        print(i)
```

```python
#Câu 3.A:
sum = 0
for i in range(11):
    if i%2 != 0:
       sum += i
print("Tổng các số: ",sum) 
```

```python
#Câu 3.C:
sum = 0
for i in range(7):
    sum += i
print("Tổng các số từ 1-6: ",sum) 
```

```python
#Câu 4:
mydict = {"a": 1, "b": 2, "c": 3, "d": 4}
```

```python
#Câu 4.A:
for key in mydict:
    print(key)
```

```python
#Câu 4.B:
for value in mydict.values():
    print(value)
```

```python
#Câu 4.C:
for key, value in mydict.items():
    print(key, value)
```
