```python
#Câu 1: Viết chương trình tính tổng 5 số được nhập vào từ bàn phím.
sum = 0
for i in range(5):
    a = input()
    sum += int(a)
print(sum)
```

```python
#Câu 2: Viết chương trình tính tổng n số được nhập vào từ bàn phím, 
#       kết thúc bằng phím ‘e’ hoặc ‘E’.
sum = 0
a = input()
while a != 'e' and a != 'E':
    if a == 'e' or a == 'E':
        pass
    else:
        sum += int(a)
        a = input()
print(sum)
```

```python
#Câu 3: Viết chương trình tính tổng các số chẳn dương được nhập vào từ bàn phím.
#       Khi chạy chương trình: xuất ra các số vừa nhập vào, và kết quả.
sum = 0
a = input()
while a != 'e' and a != 'E':
    if a == 'e' or a == 'E':
        pass
    else:
        if int(a) > 0 and int(a)%2==0:
            sum += int(a)
        a = input()
print(sum)
```

```python
#Câu 4:
n = input("Nhập n: ")
sum = 0
for i in range(int(n)):
    sum += i
print("Kết quả: " + str(sum))
```

```python
#Câu 5:
a = int(input("Nhập số thứ 1:"))
b = int(input("Nhập số thứ 2:"))
c = int(input("Nhập số thứ 3:"))
print("Số lớn nhất là: ",str(max(a,b,c)))
print("Số nhỏ nhất là: ",str(min(a,b,c)))

# Lưu ý: không dùng min, max.
```

```python
#Câu 10:
s1 = "dữ liệu"
text = "Big data is important. Data science uses data."

target = "data"
replacement = "dữ liệu"

result = ""
i = 0

while i < len(text):
    if text[i:i+len(target)].lower() == target:
        result += replacement
        i += len(target)  
    else:
        result += text[i]
        i += 1

print("Chuỗi sau khi thay thế:")
print(result)

# Lưu ý: Không dùng replace.
```

```python
#Câu 11: Cho 2 chuỗi bất kỳ. Viết chương trình xuất ra những từ vừa có ở trong chuỗi 1 
#        và chuỗi 2.
t1 = input("Nhập chuỗi thứ 1:")
t2 = input("Nhập chuỗi thứ 2:")

s1 = t1.split()
s2 = t2.split()
n = 0

for i in range(len(s1)):
    for j in range(len(s2)):
        if s1[i] == s2[j]:
            for k in range(i):
                if s1[k] == s1[i]:
                    n += 1
            if n == 0:
                print(s1[i])
            n = 0


# Lưu ý: quét qua so sánh, không xử dụng hàm.
```

```python
#Câu 12:
dssv =  {
    "MSV":[
        "9253"
        ,"7034"
        ,"6198"
    ],
    "Hoten":[
        "Đặng Thị Lan Anh"
        ,"Nguyễn Thị Phương Anh"
        ,"Trần Anh Tài"
    ],
    "NamSinh":[
        2000
        ,2002
        ,2001
    ],
    "GioiTinh":[
        "Nữ"
        ,"Nữ"
        ,"Nam"
    ],
    "DanToc":[
        "Hoa"
        ,"Kinh"
        ,"Kinh"
    ]
}

# Lưu ý: lưu bằng dictionary. Tìm theo MSV.
```

```python
mssv = input("Nhập MSV muốn tìm:")

if mssv in dssv["MSV"]:
    index = dssv["MSV"].index(mssv)
    print("Năm sinh:", dssv["NamSinh"][index])
else:
    print("Không tìm thấy sinh viên!")
```

```python
nam = 0
nu = 0

for i in dssv["GioiTinh"]:
    if i == "Nữ":
        nu += 1
    else:
        nam += 1
print("Số học sinh nam: ",nam)
print("Số học sinh nu: ",nu)
```

```python
mssv = input("Nhập MSV muốn tìm:")

if mssv in dssv["MSV"]:
    index = dssv["MSV"].index(mssv)
    for key in dssv:
        dssv[key].pop(index)
    print("Đã xoá thành công!")
else:
    print("Không tìm thấy sinh viên!")
```

```python
mssv = input("Nhập mssv: ")
if mssv in dssv["MSV"]:
    print("MSV đã tồn tại!")
else:
    name = input("Nhập họ và tên: ")
    namsinh = input("Nhập năm sinh: ")
    gioitinh = input("Nhập giới tính: ")
    
    dssv["MSV"].append(mssv)
    dssv["Hoten"].append(name)
    dssv["NamSinh"].append(namsinh)
    dssv["GioiTinh"].append(gioitinh)
    print("Đã thêm thành công!")
```
